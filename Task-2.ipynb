```python
import pandas as pd
```

```python
df= pd.read_csv("country_wise_latest.csv")
```

```python
df.head()
```

### filter data based on conditions

```python
filtered_df = df[df['Active']>10000]
print(filtered_df)
```

### calculate summary statistics

```python
summary_statistics = df.describe()
print(summary_statistics)
```

### addtional summary statistics

```python
df['Active']
```

```python
mean = df['Active'].mean()
print(mean)

median = df['Active'].median()
print(median)

standard_deviation = df['Active'].std()
print(standard_deviation)
```

### handle missing values

```python
df_filled = df.fillna(df['Active'].mean())
print("filled rows & cols")
print(df_filled)

df_dropped = df.dropna()
print("dropped rows & cols")
print(df_dropped)
```
