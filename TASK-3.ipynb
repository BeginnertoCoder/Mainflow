```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'Category': ['A', 'B', 'C'], 'Value1': [10, 15, 20], 'Value2': [5, 8, 12]}
df = pd.DataFrame(data)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

## BAR CHART

ax1.bar(df['Category'], df['Value1'], color='skyblue')
ax1.set_xlabel('Category')
ax1.set_ylabel('Value 1')
ax1.set_title('Bar Chart - Value 1')

## LINE CHART

ax2.plot(df['Category'], df['Value2'], marker='o', linestyle='-', color='coral')
ax2.set_xlabel('Category')
ax2.set_ylabel('Value 2')
ax2.set_title('Line Chart - Value 2')
plt.legend(df.columns[1:]) 

plt.tight_layout()
plt.show()
```
